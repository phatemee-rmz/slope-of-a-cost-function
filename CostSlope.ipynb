```python
import matplotlib.pyplot as plt
import numpy as np
 
%matplotlib inline
```

# Example 1 - A simple cost function
 $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
# plot
plt.xlim([-3, 3])
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(X)', fontsize = 16)
plt.plot(x_1, f(x_1))
plt.show()
```

# Slope & Derivatives
Create a python function for the derivatve of f(x) called df(x)

```python
def df(x):
    return 2*x + 1
```

```python
# plot function derivative side by side
plt.figure(figsize=[15,5])

# 1 chart: Cost Function
plt.subplot(1, 2,1)

plt.xlim([-3, 3])
plt.ylim(0, 8)

plt.title('Cost Function')
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(X)', fontsize = 16)
plt.grid()

plt.plot(x_1, f(x_1), color='purple', linewidth = 3)

# 2 chart: Derivative
plt.subplot(1,2,2)

plt.title('Slope of the cost Function')
plt.xlabel('X', fontsize = 16)
plt.ylabel('df(X)', fontsize = 16)
plt.grid()

plt.xlim([-2, 3])
plt.ylim(-3, 6) 

plt.plot(x_1, df(x_1), color='red', linewidth = 3)
plt.show()
```
